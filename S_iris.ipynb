# DECISION TREE

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier , plot_tree
```

```python
iris = pd.read_csv(r"/content/iris_cleaned.csv")

x = iris[["sepal_length","sepal_width","petal_length","petal_width"]]
y = iris[["class"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)    #Split into train test
```

```python
from sklearn import preprocessing as ppr
lab_en = ppr.LabelEncoder()
y_train['class'] = lab_en.fit_transform(y_train['class'])        # updating the class as encoded targets
y_test['class'] = lab_en.fit_transform(y_test['class'])
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([('decision_tree', DecisionTreeClassifier())], verbose = True)

#fitting training_data into pipe
pipe.fit(x_train,y_train)

decision_tree = pipe.named_steps['decision_tree']
```

```python
y_pred = pipe.predict(x_test)
y_pred
```

```python
plt.figure(figsize=(12, 8))
plot_tree(decision_tree, filled=True, feature_names=x_train.columns, class_names=True)
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))
print("\nClassification Report\n",classification_report(y_test, y_pred))
```

# SVM

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
```

```python
iris = pd.read_csv(r"/content/iris_cleaned.csv")

x = iris[["sepal_length","sepal_width","petal_length","petal_width"]]
y = iris[["class"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)    #Split into train test
```

```python
from sklearn import preprocessing as ppr
lab_en = ppr.LabelEncoder()
y_train['class'] = lab_en.fit_transform(y_train['class'])        # updating the class as encoded targets
y_test['class'] = lab_en.fit_transform(y_test['class'])
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([ ('std', StandardScaler()), ('svm', SVC())], verbose = True)

#fitting training_data into pipe
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))
print("\nClassification Report\n",classification_report(y_test, y_pred))
```
